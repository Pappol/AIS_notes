```python
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from scipy import ndimage as nd
import skimage
```

```python
img = Image.open("ballons.png")
bw = img.convert("L")
plt.imshow(bw)
plt.show()
```

```python
gaussian_img4 = nd.gaussian_filter(bw, sigma=4)
gaussian_img7 = nd.gaussian_filter(bw, sigma=7)
median_img = nd.median_filter(bw, size=5)

f, axarr = plt.subplots(nrows=1,ncols=3, figsize=(25, 18))
plt.sca(axarr[0])
plt.imshow(gaussian_img4, interpolation='nearest')
plt.title('Gaussian sigma=4')
plt.sca(axarr[1]) 
plt.imshow(gaussian_img7, interpolation='nearest')
plt.title('Gaussian sigma=7')
plt.sca(axarr[2]);
plt.imshow(median_img, interpolation='nearest')
plt.title('Median filter size=5')

plt.show()
```

```python
median_img2 = nd.median_filter(median_img, size=7)

f, axarr = plt.subplots(nrows=1,ncols=2, figsize=(15, 18))
plt.sca(axarr[0])
plt.imshow(median_img, interpolation='nearest')
plt.title('Median filter 1x5- First Iteration')
plt.sca(axarr[1]) 
plt.imshow(median_img2, interpolation='nearest')
plt.title('Median filter 1x7- Second iteration')
```
